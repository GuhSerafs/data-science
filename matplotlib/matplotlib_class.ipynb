# Line Charts

Gráficos x,y onde x é a variável de tempo e os pontos são conectados por linhas.

```python
# Importando as bibliotecas necessárias

import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x1 = np.linspace(0, 20, 1000) # Um vetor de 0..20 com 1000 pontos
y1 = np.sin(x1)+0.2*x1

plt.plot(x1, y1)
plt.xlabel("Input Data")
plt.ylabel("Output Data")
plt.title("My Transfer Function")
plt.show() # show -> necessário para caso não seja chamado dentro de um notebook
```

# Scatter Plot
Plot 2D onde os pontos são representados por pontos ou outros tipos de símbolos

```python
RandomMatrix = np.random.randn(100, 2)
plt.scatter(RandomMatrix[:, 0], RandomMatrix[:, 1])
plt.show()
```

```python
# É possível usar cores
distributionCenterDeviation = np.linspace(0, 5, 100)
RandomMatrix[:, 0] += distributionCenterDeviation
plt.scatter(RandomMatrix[:,0], RandomMatrix[:, 1], c=distributionCenterDeviation)
plt.show()
```

# Histograms
Serve para plotar a quantidade de vezes cada variável se repete

```python
HistogramRandomMatrix = np.random.randn(1000)
plt.hist(HistogramRandomMatrix)
plt.show()
```

# Image Plot
Através da biblioteca PIL é possível abrir uma imagem. 


```python
from PIL import Image
mercedesSLS = Image.open('sls-amg.jpg')
imageArray = np.array(mercedesSLS)
plt.imshow(mercedesSLS)
plt.show()
```

```python
# Analisando o conteúdo do array mercedesSLS
print(imageArray)
print(type(imageArray))
print(imageArray.shape)
# etc
```

```python
# Obtendo um 'heatmap' da imagem / ou escala de cinza 
grayScale = np.mean(imageArray, axis=2)
plt.imshow(grayScale, cmap='gray')
plt.show()
```

```python
# Comparação entre os elementos do grayscale e o original
def getPointInfo(x, y): 
    average = grayScale[x,y]
    rgb = imageArray[x, y, :]
    return average, rgb

avg, rgb = getPointInfo(120, 350)
print(avg, rgb)
```

```python
# Exercise - Implementar um XOR genérico 
N = 2000
randMatrix = 2*np.random.random([N, 2]) - 1

def generalXor(x, y): 
    if ((x > 0 and y > 0) or (x < 0 and y < 0)): 
        return 0
    else:
        return 1

points = zip(randMatrix[:, 0], randMatrix[:, 1])
colormap = np.array([generalXor(x, y) for x, y in points])
plt.scatter(randMatrix[:, 0], randMatrix[:, 1], c = colormap)
plt.title('XOR Genérico')
plt.show()
```
